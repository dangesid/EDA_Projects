```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("/content/DataNetflixRevenue2020_V2.csv")
df2 = pd.read_csv("/content/DataNetflixSubscriber2020_V2.csv")
df3 = pd.read_csv("/content/NetflixSubscribersbyCountryfrom2018toQ2_2020.csv")
df4 = pd.read_csv("/content/NetflixsRevenue2018toQ2_2020.csv")
```

```python
df = pd.concat([df1,df2],axis=1)
f_df = df.iloc[:,2:]
f_df.head()
```

```python
f_df.columns
```

```python
f_df.info()
```

```python
f_df["Quaters"] = f_df["Years"].str.split("-").str[0]
f_df["Year"] = f_df["Years"].str.split("-").str[1]
f_df = f_df.drop("Years",axis=1)
f_df.head()
```

```python
f_df["Year"].unique()


```

```python
# Plot the graph of Quaters vs Revenue

plt.figure(figsize=(8,6))
plt.bar(f_df["Year"],f_df["Revenue"])

plt.title('Year vs Revenue')
plt.xlabel('Year')
plt.ylabel('Revenue')

plt.grid(True)
plt.show()
```

```python
# Plot the graph of Quaters vs Revenue

plt.figure(figsize=(10,10))
plt.bar(f_df["Area"],f_df["Revenue"])

plt.title('Quarters vs Revenue')
plt.xlabel('Quarters')
plt.ylabel('Revenue')

plt.grid(True)
plt.show()
```
